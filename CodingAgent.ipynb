```python
!!pip install litellm

# Important!!!
#
# <---- Set your 'GEMINI_API_KEY' as a secret over there with the "key" icon
#
#
import os
from google.colab import userdata
import google.generativeai as genai

# 1. Retrieve the key from Colab Secrets
api_key = userdata.get('GEMINI_API_KEY')

# 2. Configure the Gemini SDK with the API key globally
genai.configure(api_key=api_key)

# 3. Initialize the model (Gemini 2.5 Flash is now being used, as it was listed as available.)
# The API key is now configured globally, so it's not needed here.
model = genai.GenerativeModel('gemini-2.5-flash') # Updated to an available model

# Example usage:
response = model.generate_content("Hello! Give me a quick tip for using the Gemini API.")
print(response.text)

# To see available models, you can uncomment and run the following:
for m in genai.list_models(): # No need to pass api_key if configured globally
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
from litellm import completion
from typing import List, Dict

def generate_response(messages: List[Dict]) -> str:
    """Call LLM to get response"""
    # Access the model name string from the genai.GenerativeModel object
    # The model variable is defined in the previous cell as genai.GenerativeModel('gemini-2.5-flash')
    # We need to pass the string 'gemini-2.5-flash' to litellm.
    # model.model_name returns 'models/gemini-2.5-flash'. We need to extract 'gemini-2.5-flash'.
    model_id = model.model_name.split('/')[-1]
    response = completion(
        model=f"gemini/{model_id}", # Pass the model ID without 'models/' prefix, with 'gemini/' prefix for litellm
        messages=messages,
        max_tokens=4096, # Increased max_tokens to allow for a more complete response
        api_key=api_key # Pass the API key explicitly to litellm
    )
    return response.choices[0].message.content


what_function_help_with = input("What do you want code for?")

messages = [
    {"role": "system", "content": "You are a helpful software engineer. You will ask the user what they want code for, write the code for the function, add documentation, and finally include test cases using the unittest framework."},
    {"role": "user", "content": what_function_help_with}, # Corrected variable name and added comma

   # Here is the assistant's response from the previous step
   # with the code. This gives it "memory" of the previous
   # interaction.
   {"role": "assistant", "content": response}, # Added comma

   # Now, we can ask the assistant to update the function with documentation
   {"role": "user", "content": "Update the function to include documentation that covers: Function description, Parameter descriptions, Return value description, Example usage, Edge cases."}, # Added comma

   # Here is the assistant's response from the previous step
   # with the code. This gives it "memory" of the previous
   # interaction.
   {"role": "assistant", "content": response}, # Added comma

   # Now, we can ask the assistant to update the function and include test cases using Pythons unittest framework
   {"role": "user", "content": "Update the function to include test cases using Pythons unittest framework. Tests should cover: Basic functionality, Edge cases, Error cases,Various input scenarios"}

]

response = generate_response(messages)
print(response)
```
